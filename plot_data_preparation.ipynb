This notebook creates the data in JavaScript format needed to plot circular connection plots.<br>
It requires the input to be exactly the same as <i>OnlySigData_IQ_4Dec2020.xlsx</i> to work.<br>
Once the js codes are generated, copy and paste them into the appropriate location in the html file.

```python
# Load data
folder = 'C:\\Users\\Sen\\Desktop\\LMU_graphic_design\\Rachele_paper_plot\\circular_plot\\'
excel_file = 'OnlySigData_IQ_4Dec2020.xlsx'
# Define the subpages in the input excel
sheets = ['PSC2_lh','PSC5_lh','PSC6_lh','PSC7_lh','PSC12_lh','PSC1_rh','PSC3_rh','PSC5_rh','PSC9_rh','PSC10_rh','PSC12_rh','PSC13_rh','PSC16_rh','PSC18_rh']
# Define the features in each of the subpages to be plotted
feature_names = ['PSC 2 left','PSC 5 left','PSC 6 left','PSC 7 left','PSC 12 left','PSC 1 right','PSC 3 right','PSC 5 right','PSC 9 right','PSC 10 right','PSC 12 right','PSC 13 right','PSC 16 right','PSC 18 right']
# Define the names of different colors to be used in the plot
color_names = ['PSC_2_left','PSC_5_left','PSC_6_left','PSC_7_left','PSC_12_left','PSC_1_right','PSC_3_right','PSC_5_right','PSC_9_right','PSC_10_right','PSC_12_right','PSC_13_right','PSC_16_right','PSC_18_right']

xls = pd.ExcelFile(folder+excel_file)
output_commands = []
for s,f in zip(sheets,feature_names):
    df = pd.read_excel(xls,s)
    df = df.dropna()
    df['output_command'] = '{"from":"' + f + '","to":"' + df['Specific Item'].str.strip() + '","value":' + df['Unnamed: 4'].astype(str) + ',"r_value":' + df['r-value'].astype(str) + '},'
    output_commands += list(df['output_command'])
```

Examine the processed dataframe. The output_command column is relevant to plot creation.

```python
for line in output_commands[::-1]:
    # These are customisations made specifically with the given data file, not generalisable
    line = line.replace('highest','high')
    line = line.replace('lowest','low')
    line = line.replace('highest lifetime','lifetime')
    print (line)
```
